```python
import re
from glob import glob
from functools import reduce

parcel_paths = [p for p in sorted(glob('data/MinneMUDAC_raw_files/*.txt', recursive=True)) if 'parcels' in p]
```

```python
year_pattern = re.compile(r'(\d{4})_metro_tax_parcels')
```

```python
getyear = lambda path: (n.group(1) if (n := year_pattern.search(path)) else None)
```

```python
years = [getyear(path) for path in parcel_paths]

print(years)
```

```python
(schemas := {
    path: {col: pl.String for col in pl.scan_csv(path,has_header=True, ignore_errors=True, separator='|').limit(1)
           .collect().columns}
    for path in parcel_paths
}) #columns to keep
```

# New Section

```python
(indicator_tables := [
    (
    pl.scan_csv(path, has_header=True, separator='|',schema=schemas[path]).limit(1)
    .unpivot()
    .rename({'variable': 'column'})
    .drop('value')
    .with_columns(pl.lit(1).alias(getyear(path)))
    .collect()
) for path in parcel_paths])
```

```python
join_next = lambda left, right: left.join(right, on='column',how='full')
```

```python
join_next(indicator_tables[0], indicator_tables[1])
```

```python
def join_next(left, right):
    year = right.columns[-1]  # or however you track the year
    return left.join(right, on="column", how="inner", suffix=f"_{year}")
```

```python
(indicator_table := (
    reduce(join_next, indicator_tables)
 .filter(pl.col('column').is_not_null())
 .fill_null(0)
))
```

```python
indicator_table.select(cs.integer()).sum()
#remove 2002 and 2003 because it is missing a lot of data
```
